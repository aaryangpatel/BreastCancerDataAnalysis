```python
# Breast Cancer Analysis with Decision Tree Classification using a Variety of Pre-Prunning Methods
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import export_graphviz
import graphviz
```

```python
# Load the data
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=80)
```

```python
# Create a pure decision tree
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
# Create a visualization of the decision tree using graphviz
export_graphviz(tree, out_file="tree.dot", class_names=["malignant", "benign"], 
                                feature_names=cancer.feature_names, impurity=False, filled=True)

with open("tree.dot") as f: dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
# A function to plot the importance (0 to 1) of each feature in the decision tree's result
def plot_feature_importances(model):
    n_features = cancer.data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center') 
    plt.yticks(np.arange(n_features), cancer.feature_names) 
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
```

```python
# Parameter range to test
max_depths = range(1,10)
min_samples_splits = range(2,20)

max_depth_results = []
min_samples_split_results = []
```

```python
# Test the effect of different max depth values on accuracy
for max_depth in max_depths:
    tree = DecisionTreeClassifier(max_depth=max_depth, random_state=0)
    tree.fit(X_train, y_train)
    test_acc = accuracy_score(y_test, tree.predict(X_test))
    max_depth_results.append(test_acc)
    plot_feature_importances(tree)
```

```python
# Test the effect of different min samples split values on accuracy
for min_samples_split in min_samples_splits:
    tree = DecisionTreeClassifier(min_samples_split=min_samples_split, random_state=0)
    tree.fit(X_train, y_train)
    test_acc = accuracy_score(y_test, tree.predict(X_test))
    min_samples_split_results.append(test_acc)
```

```python
# Plot accuracy vs. max_depth
plt.plot(max_depths, max_depth_results, label='Test Accuracy')
plt.xlabel('Max Depth')
plt.ylabel('Test Accuracy')
plt.title('Effect of Max Depth on Test Accuracy')
```

```python
# Plot accuracy vs. min_samples_split
plt.plot(min_samples_splits, min_samples_split_results, label='Test Accuracy')
plt.xlabel('Min Samples Split')
plt.ylabel('Test Accuracy')
plt.title('Effect of Min Samples Split on Test Accuracy')
```
